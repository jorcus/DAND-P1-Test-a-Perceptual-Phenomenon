## Background Information 
In a Stroop task, participants are presented with a list of words, with each word displayed in a color of ink. The participant’s task is to say out loud the color of the ink in which the word is printed. The task has two conditions: a congruent words condition, and an incongruent words condition. In the congruent words condition, the words being displayed are color words whose names match the colors in which they are printed: for example RED, BLUE. In the incongruent words condition, the words displayed are color words whose names do not match the colors in which they are printed: for example PURPLE, ORANGE. In each case, we measure the time it takes to name the ink colors in equally-sized lists. Each participant will go through and record a time from each condition.

### 1. What is our independent variable? What is our dependent variable?
Independent variable is a variable (often denoted by x ) whose variation does not depend on that of another.

Dependent variable is a variable (often denoted by y ) whose value depends on that of another.

DEPENDENT   - The Response Time to name the ink colors in equally-sized lists of congruent and incongruent words.

### 2. What is an appropriate set of hypotheses for this task? What kind of statistical test do you expect to perform? Justify your choices.

μC - Population mean of Congruent Test

μI - Population mean of Incongruent Test

Ho - **Null Hypothesis**: (  μi - μc = 0 || μi = μc ) The population average response time in viewing the congruent(c) words and viewing the incongruent(i) words are equal.

Ha - **Alternative Hypothesis**: ( μi - μc ≠ 0 ||  μi > μc || μC ≠ μI) Population mean of Incongruent Test will take longer than congruent test.

Since there is only one independent variable with 2 levels, I think that t-Test is the most appropriate statistical test. Also, It's a two tailed t-test for this task. 

### 3. Report some descriptive statistics regarding this dataset. Include at least one measure of central tendency and at least one measure of variability.

```python
meanA = np.mean(dframe['Congruent']); round(meanA,2)
stdA = np.std(dframe['Congruent'])
medianA = np.median(dframe['Congruent'])
rangeA = dframe['Congruent'].max() - dframe['Congruent'].min()
varA = np.var(dframe['Congruent'])


meanB = np.mean(dframe['Incongruent']); round(meanA,2)
stdB = np.std(dframe['Incongruent'])
medianB = np.median(dframe['Incongruent'])
rangeB = dframe['Incongruent'].max() - dframe['Incongruent'].min()
varB= np.var(dframe['Incongruent'])


print('Congruent (Mean): ' + str(round(meanA,4)))
print('Congruent (Standard Deviation): ' + str(round(stdA,4)))
print('Congruent (Median): ' + str(round(medianA,4)))
print('Congruent (Range): ' + str(round(rangeA,4)))
print('Congruent (Variance): ' + str(round(varA,4)) + '\n')

print('Incongruent (Mean): ' + str(round(meanB,4)))
print('Incongruent (Standard Deviation): ' + str(round(stdB,4)))
print('Incongruent (Median): ' + str(round(medianB,4)))
print('Incongruent (Range): ' + str(round(rangeB,4)))
print('Incongruent (Variance): ' + str(round(varB,4)))
dframe.describe()
```

```python
n=len(dframe)

#point estimate of μi - μc -> PE (in seconds)
PE = meanB - meanA
round(PE,2)

#s - sample standard deviation of differences
#1. D = sample difference
dframe['Difference'] = dframe['Incongruent'] - dframe['Congruent']

#2. DFM = difference from the mean
#SQD = squared differences from the mean
DFM = dframe['Difference'] - dframe['Difference'].mean()
dframe['SQD'] = DFM*DFM

#3. SSD = sum of squared differences
SSD = dframe['SQD'].sum()

#4. v = variance = SSD/(n-1)
v = SSD/(n-1)

#5. s = sqrt(v)
s = sqrt(v)
round(s,2)

#t-statistic
#t = PE/(s/√n)
#PE = (μi - μc)

t = PE/(s/(sqrt(n)))
round(t,4)

dframe.head()
```
